```python
import pandas as pd
import math
```

```python
df = pd.read_csv("cleaned_data.csv")
```

```python
df
```

```python
BCR_unique = df['BCR'].unique()
```

```python
bcr_to_state = pd.read_csv("/content/BCR_States_Coverage.csv")
```

```python
bcr_to_state
```

```python
# Merge bird data with BCR-to-state data
merged_data = pd.merge(df, bcr_to_state, left_on="BCR", right_on="BCR Name")

# Calculate the BirdCount for each state based on coverage percentage
merged_data["NumBird"] = ((merged_data["BirdCount"] * merged_data["Coverage"]) / 100).round().astype(int)

# Drop unnecessary columns for a cleaner result
final_data = merged_data.drop(columns=["BCR Name", "Coverage", "BCR", "BirdCount"])
```
